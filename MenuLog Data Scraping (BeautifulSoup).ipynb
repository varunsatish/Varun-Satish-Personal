```python
import pandas as pd
import urllib
from bs4 import BeautifulSoup
import lxml
```

```python
redfern = "https://www.menulog.com.au/area/2016-redfern/?lat=-33.89334&long=151.20461"
blacktown = "https://www.menulog.com.au/area/2148-blacktown/?lat=-33.771&long=150.9063"
newtown = "https://www.menulog.com.au/area/2042-newtown/?lat=-33.897&long=151.1793"
page = urllib.request.urlopen(newtown)
soup = BeautifulSoup(page, "lxml")
```

```python
soup.title #title of data source
title = soup.title.string #returns string of title
```

```python
#h3 is the class attached to all the names of the restaurants

restaurants_raw = soup.find_all("h3")
address_raw = soup.find_all(itemprop = "address")
cuisine_raw = soup.find_all(itemprop = "servesCuisine") 
agg_rating_raw = soup.find_all(itemprop = "aggregateRating")
rating_raw = soup.find_all(itemprop = "ratingValue")
```

```python
def finder(cat,raw):
    "getting text from raw HTML data"
    for element in raw:
        cat[element.get(text())] = {}
```

```python
newtown_diff = len(restaurants_raw)-len(rating_raw)
```

```python
rest = {}
add = {}
cui = {}

import re as re

rating_string = str(rating_raw)

rat = re.findall("\d+\.+\d", rating_string)  #this extracts the raw digits of the ratings from using regular expressions

agg_rat = {}
```

```python
#for each element in our raw resturant data, we are extracting the text

for element in restaurants_raw:
    rest[element.get_text()] = {}

for element in address_raw:
    add[element.get_text()] = {}    
    
for element in cuisine_raw:
    cui[element.get_text()] = {}
        
for element in agg_rating_raw:
     agg_rat[element.get_text()] = {}
```

```python
#getting dict class into dataframe 
restaurant_list = pd.DataFrame(list(rest.items()))
restaurant_list.columns = ['Name', 'b']

dropper = list(range(len(restaurant_list) - newtown_diff,len(restaurant_list)))
restaurant_list = restaurant_list.drop(dropper)

address_list = pd.DataFrame(list(add.items()))
address_list.columns = ['Address', 'b']

dropper = list(range(len(address_list) - newtown_diff,len(address_list)))
address_list = address_list.drop(dropper)

cuisine_list = pd.DataFrame(list(cui.items()))
cuisine_list.columns = ['Cuisine', 'b']

dropper = list(range(len(cuisine_list) - newtown_diff,len(cuisine_list)))
cuisine_list = cuisine_list.drop(dropper)

rating_list = pd.DataFrame({'Rating': rat})

agg_rating_list = pd.DataFrame(list(agg_rat.items()))
agg_rating_list.columns = ['Aggregate Rating', 'b']
```

```python
data_newtown = pd.DataFrame(restaurant_list['Name'])
```

```python
data_newtown['Name'] = restaurant_list['Name']
data_newtown['Address'] = address_list['Address']
data_newtown['Rating'] = rating_list['Rating']
data_newtown['Cuisine'] = cuisine_list['Cuisine']
data_newtown['Aggregate Rating'] = agg_rating_list['Aggregate Rating']
```
